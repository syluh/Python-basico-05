### Desarrolle el Reto

```python
from ipyleaflet import Map, MagnifyingGlass, basemaps, basemap_to_tiles, Marker
import pandas as pd
import ipywidgets as widgets
```

```python
datosU =  pd.read_json("data/ocw_universidades.json")
```

```python
diccionarioU = datosU.to_dict("records")
diccionarioU 
```

```python
lista_tuplas_U = [((d['universityName']), (d['universityLat'], d['universityLong'],d['universityName'] )) 
                  for d in diccionarioU ]
```

```python
lista_tuplas_U[0]
```

```python
lat_long = widgets.Dropdown(
    options = lista_tuplas_U,
    value=(-1.294246, 36.786794, 'African Virtual University'),
    description= 'Universidad:',
)
lat_long
```

```python
center = (0,0)
m = Map(center=center, zoom = 2, basemap=basemaps.Stamen.Terrain)

if(lat_long.value[0]is not None) and (lat_long.value[1]is not None):
    c = (lat_long.value[0],lat_long.value[1])
    marker = Marker(location=c, draggable = False,
                   title="%s"%(lat_long.value[2]),alt=lat_long.value[2])
    m.add_layer(marker);
    print(lat_long.value[2])
m
```
